```python
import random, json, time
from datetime import datetime


def generate_key():

    return bytes([random.randint(0, 255) for _ in range(16)])

def aes_encrypt(plain_bytes, key):
    """Using XOR """
    return bytes([b ^ key[i % len(key)] for i, b in enumerate(plain_bytes)])

def aes_decrypt(cipher_bytes, key):
    """Decryption (XOR reversible)"""
    return aes_encrypt(cipher_bytes, key)  # نفس العملية


def read_sensor():
    temp = round(random.uniform(20, 40), 2)
    hum  = round(random.uniform(30, 70), 2)
    return {"temp": temp, "hum": hum, "time": datetime.now().isoformat()}


def simulate():
    data = read_sensor()
    print("Original Data:", data)

    key = generate_key()
    plain_bytes = json.dumps(data).encode('utf-8')

    cipher = aes_encrypt(plain_bytes, key)
    print("Encrypted Data (hex):", cipher.hex()[:30] + "...")

    time.sleep(0.5)

    decrypted = aes_decrypt(cipher, key)
    recovered = json.loads(decrypted.decode('utf-8'))
    print("Data After Decryption:", recovered)

simulate()
```
